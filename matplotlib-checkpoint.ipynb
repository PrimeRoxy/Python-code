```python
from matplotlib import pyplot as plt
subject = ['english', 'maths', 'hindi', 'science', 'computer']
Marks = [56, 78, 80, 76, 99]
Marks1 = [77,45, 66, 88, 76]
plt.scatter(subject,Marks, label='R', color='red', marker='*')
plt.bar(subject,Marks1, label= 's')
plt.title('Results')
plt.ylabel('M')
plt.legend()
plt.show()
```
